```python
import csv
import json
```

```python
filesys_path  = 'files/'
file_data_in  = 'Data_Scores_aznetz_Details_20200430.txt'
file_json_out = 'aznetz_Details_20200430.json'
file_csv_out  = 'aznetz_Details_20200430.csv'
```

```python
def add2matrix(matrix,key,value):
    try:
        matrix[key] = matrix[key]+[value]
    except KeyError:
        matrix[key] = []
    return matrix
```

```python
def initialize():
    vars = {
    "portals" : ["AAZ","BAT","BLZ","BZB","GTB","LIZ","OT","SOZ"],
    "portals_i" : 0,
    "flag" : 0,
    "current_portal" : '',
    "current_section" : '',
    "current_key" : '',
    "matrix" : {}
    }
    return vars
```

```python
def insert_empty_row(matrix,ptl,section):
    #print(matrix)
    key = ' '.join([ptl,section])
    #print("ptl: %s, section: %s key: %s" % (ptl,section,key))
    #print(key)
    for i in range(0,6):
        matrix = add2matrix(matrix,key,0)
    return matrix
```

```python
vars = initialize()
with open(filesys_path+file_data_in,'r',encoding="utf-8") as csvfile:
    reader = csv.reader(csvfile, delimiter=';', quotechar='"') # Delimiter darf nicht Komma sein!
    for row in reader:
        try:
            row[0]
            if row[0]=="Summe":
                if vars["flag"] == 0:
                    vars["current_portal"] = vars["portals"][vars["portals_i"]]
                    vars["matrix"] = insert_empty_row(vars["matrix"],vars["current_portal"],"APP-EPAPER")
                    vars["current_section"] = 'APP-LIVE'
                    vars["flag"] = 1
                    vars["portals_i"] = vars["portals_i"] + 1
                elif vars["flag"] == 1:
                    vars["flag"] = 0
                    vars["current_section"] = 'WEB'
                else:
                    pass
            vars["current_key"] = ' '.join([vars["current_portal"],vars["current_section"]])
            vars["matrix"] = add2matrix(vars["matrix"],vars["current_key"],row[0].replace(',',''))
        except IndexError:
            pass
print(vars["matrix"])
```

```python
f = open(filesys_path+file_json_out,"w")
f.write(json.dumps(vars["matrix"]))
f.close()
```
